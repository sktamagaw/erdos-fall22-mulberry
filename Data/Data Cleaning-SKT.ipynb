In this notebook, we open the data set downloaded from <a href="https://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients#"> this website</a>, and resave it with the column titles as the column headers

```python
#Importing packages
import pandas as pd
```

```python
#Loading the data set as data
data=pd.read_excel("raw_data.xls")
```

```python
data.head()
```

Checking to see if there is any missing data:

```python
X=[]
for i in range(1,24):
    X.append("X"+str(i))
```

```python
columns = ['Unnamed: 0']+ X+["Y"]
```

```python
for x in X:
    for i in range(1,3001):
        if type(data[x][i]) != int:
            print(x)
            print(i)
            print(type(data[x][i]))
```

There don't seem to be any missing entries.  It feels more natural to have the columns headers be the titles, instead of the first row, so I'll fix that here:

```python
#make a new dataframe
newcols=data.iloc[0]
new_df = pd.DataFrame()
for c in columns:
    new_df[data[c][0]]= data[c].iloc[1:]
```

```python
new_df.head()
```

```python
#Saving the new dataframe as clean_data.csv
new_df.to_csv("clean_data.csv", index=False)
```
